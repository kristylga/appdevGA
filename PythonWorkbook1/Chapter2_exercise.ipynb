```python
Write a program to prompt the user for hours and rate per hour to compute gross pay.

Enter Hours: 35
Enter Rate: 2.75
Pay: 96.25

# Prompt the user to enter hours worked
hours = float(input("Enter Hours: "))

# Prompt the user to enter the rate per hour
rate = float(input("Enter Rate: "))

# Calculate the gross pay
pay = hours * rate

# Print the calculated gross pay
print("Pay:", pay)
```

```python
Write a program that uses input to prompt a user for a number and then outputs the square of the number

Enter a number: 3
The square of 3 is 9

# Prompt the user to enter a number
number = float(input("Enter a number: "))

# Calculate the square of the number
square = number ** 2

# Print the result
print("The square of", number, "is", square)
```
